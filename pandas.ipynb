- data handling and data deletion two ways to handle the null values in a dataset 
- data handling agr column important h to 
- data deletion agar column imp nii h 

- handling : fillna
- deletion: dropna for deletion of rows of null values
- deletion: drop for deleting the whole colulmn

```python
df['age'].fillna(df['age'].mean(),inplace=True) # fill null values in 'age' with mean
```

Filtering Data

```python
adults = df[df['age']>=18]
females = df[df['sex']=='female']
survivors = df[df['survivor']==1]
print(adults)
print(females)
print(survivors)
```
